```python
import pandas as pd
```

```python
data = pd.read_pickle("data/gpt4_choice_words_images.pkl")
data
```

```python
def compute_accuracy(data):
    n = len(data)
    human_ooo = data["human_ooo"]
    image_ooo = data["gpt_image_ooo"]
    word_ooo = data["gpt_word_ooo"]

    human_image = sum(human_ooo == image_ooo) / n 
    human_word = sum(human_ooo == word_ooo) / n
    image_word = sum(image_ooo == word_ooo) / n

    print("Accuracy human-image: ", human_image * 100)
    print("Accuracy human-word: ", human_word * 100)
    print("Accuracy image-word: ", image_word * 100)

```
